```python
import numpy as np

print(np.exp([1,2,3,4]))
print(np.sqrt([1,2]))
print(np.square([3,4]))
print(np.cbrt([64]))
print(np.reciprocal([1,2,4,5]))
print(np.sign([-10,0,10]))
print(np.abs([-1,-2,3,-4]))
print(np.fabs([-1.5,-2.3,3.2]))
print(np.ceil([1.2,2.3,3]))
print(np.rint([1.2,2.6,3]))
print(np.trunc([1.8,2.9,3]))

print(np.log10([1,10]))
print(np.log2([1,2,4,8]))
print(np.expm1([1,2,3]))
print(np.log1p([1,2,3]))
print(np.degrees([np.pi/2, np.pi]))
print(np.radians([90,180]))
print(np.sin(np.pi/2))
print(np.cos(np.pi))
print(np.tanh([1,2,3]))

print(np.add([1,2,3],[4,5,6]))
print(np.subtract([10,20,30],[1,2,3]))
print(np.multiply([1,2,3],[4,5,6]))

print(np.divide([10,20,30],[2,4,5]))
print(np.power([2,3,4],[2,2,2]))
print(np.maximum([1,5,3],[2,4,6]))
print(np.minimum([1,5,3],[2,4,6]))
print(np.fmax([1,np.nan,3],[2,4,6]))
print(np.fmin([1,np.nan,3],[2,4,6]))
print(np.mod([10,20,30],[1,2,3]))
print(np.remainder([10,20,30],[1,2,3]))
```

```python
import numpy as np

arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))

print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))

print(np.percentile(arr, 50))
print(np.quantile(arr, 0.5))
print(np.average(arr))
print(np.nanmean([1,2,np.nan,4]))
print(np.nansum([1,2,np.nan,4]))
print(np.nanmin([1,2,np.nan,4]))
print(np.nanmax([1,2,np.nan,4]))
print(np.sum([[1,2],[3,4]], axis=0))
print(np.sum([[1,2],[3,4]], axis=1))
```

```python
#Array Manipulation
import numpy as np

x = [[1,2,3], [3,4,5]]
arr = np.array(x)
print(arr)
arr1 = np.reshape(arr, (3,2))
arr2 = np.ravel(arr) #change to 2d to 1d
arr3 = np.transpose(arr) #changes rows->column viseversa
print(arr1)
print(arr2)
```
